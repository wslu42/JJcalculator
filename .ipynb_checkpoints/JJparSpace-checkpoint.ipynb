```python
from mpl_toolkits import mplot3d
def viewGraph (ax, view = '45'):
    import operator
    ops = { "+": operator.add, "-": operator.sub } # etc.
    ax.view_init(15, 45)
    if view[1] == 't':
        ax.view_init(0, ops[view[0]](90,90))
    elif view[1] == 'r':
        ax.view_init(0,  ops[view[0]](0,1)*-90)
    elif view[1] == 'e':
        ax.view_init(90, ops[view[0]](90,90))
```
